```python
import pandas as pd
import pickle 
import cv2

DATA_ROOT = "C:/Users/marko/Desktop/NAIST-UBI-RESEARCH/FSAR_Team_Sports/multisports_data/data/trainval"
MEDATATA_PATH = DATA_ROOT + "/multisports_GT.pkl"

with open(MEDATATA_PATH, "rb") as file:
    data = pickle.load(file)

labels = data["labels"]
keys = list(data["nframes"].keys())

rand_sample_name = keys[654]
rand_sample_vid = cv2.VideoCapture(DATA_ROOT + f"/{rand_sample_name}.mp4")
rand_sample_tube = data["gttubes"][rand_sample_name][21][0]

rand_sample_tube_frames = rand_sample_tube[:,0]
print(labels[21])
```

# VISUALIZATION OF GROUND TRUTH

```python
i = 0

while rand_sample_vid.isOpened():
    ret, frame = rand_sample_vid.read()

    if not ret:
        break    

    if i in rand_sample_tube_frames:
        x1,y1, x2, y2 = rand_sample_tube[i - int(rand_sample_tube[0,0])][1:5].astype(int)
        cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)


    cv2.imshow("Video",frame)

    if cv2.waitKey(25) & 0xFF == ord("q"):
        break

    i += 1

rand_sample_vid.release()
cv2.destroyAllWindows()
```
